```python
fig_df2 = px.line(df2, x="obsDt", y="howMany", color="comName", title="Observaciones de aves en el tiempo")
fig_df2.update_traces(mode="lines+markers")
```

```python


mapa_calor = folium.Map(location=[df2['lat'].mean(), df2['lng'].mean()], zoom_start=10)


heat_data = df2[['lat', 'lng', 'howMany']].dropna().values.tolist()


HeatMap(heat_data).add_to(mapa_calor)


mapa_calor
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df2['howMany'], bins=30, kde=True)
plt.title('Distribución de la cantidad de aves observadas')
plt.xlabel('Cantidad de aves observadas')
plt.ylabel('Frecuencia')
plt.show()
```

```python
mapa_aves = folium.Map(location=[df2['lat'].mean(), df2['lng'].mean()], zoom_start=10)

for lat, lng, how_many in heat_data:
    folium.CircleMarker(
        location=[lat, lng],
        radius=how_many / 10,  
        color='blue',
        fill=True,
        fill_color='blue',
        fill_opacity=0.6,
        popup=f'Cantidad: {how_many}'
    ).add_to(mapa_aves)

mapa_aves
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df2, x='howMany', palette='viridis')
plt.title('Conteo de aves observadas por cantidad')
plt.xlabel('Cantidad de aves observadas')
plt.ylabel('Conteo')
plt.xticks(rotation=45)
plt.show()
```
